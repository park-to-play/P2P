테이블 별 데이터 확인 (상위20개)

```python
import pymysql
import pandas as pd

# 데이터베이스 연결 설정
connection = pymysql.connect(
    host=,
    port=,
    user=,
    password=,
    database=,
    charset=
)

# 테이블 목록
tables = ['tb_address', 'tb_parking', 'tb_traffic']

try:
    with connection.cursor() as cursor:
        for table in tables:
            print(f"테이블 '{table}' 상위 20개")
            
            query = f"SELECT * FROM {table} LIMIT 20;"
            cursor.execute(query)
            rows = cursor.fetchall()
            
            columns_query = f"SHOW COLUMNS FROM {table};"
            cursor.execute(columns_query)
            columns = [column[0] for column in cursor.fetchall()]
            
            df = pd.DataFrame(rows, columns=columns)
            print(df)
except Exception as e:
    print(f"오류{e}")

finally:
    connection.close()
```

```python
import pandas as pd
from sqlalchemy import create_engine

# SQLAlchemy 엔진 생성
connection_string = 'mysql+pymysql://root:?charset=utf8'
engine = create_engine(connection_string)

try:
    df_address = pd.read_sql("SELECT * FROM tb_address;", engine)
    df_parking = pd.read_sql("SELECT * FROM tb_parking;", engine)
    df_traffic = pd.read_sql("SELECT * FROM tb_traffic;", engine)

    print("tb_address")
    print(df_address.head())

    print("tb_parking")
    print(df_parking.head())

    print("tb_traffic")
    print(df_traffic.head())

except Exception as e:
    print(f"오류: {e}")

finally:
    # 엔진 종료
    engine.dispose()
```
