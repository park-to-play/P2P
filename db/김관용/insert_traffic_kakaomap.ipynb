```python
create_table_query="""
CREATE TABLE IF NOT EXISTS tb_traffic (
    traffic_id INTEGER NOT NULL AUTO_INCREMENT,
    name VARCHAR(50) NULL,
    congestion INTEGER NULL,
    distance VARCHAR(50) NULL,
    firstUpdateTime VARCHAR(14) NULL,
    speed INTEGER NULL,
    direction VARCHAR(50) NULL,
    coordinates TEXT NULL,
    PRIMARY KEY (traffic_id)
);
"""
```

```python
connection = pymysql.connect(
    host=,
    port=,
    user=,
    password=,
    database=,
    charset=
)
connection

try:
    with connection.cursor() as cursor:
        # 테이블 생성
        cursor.execute(create_table_query)
        connection.commit()
        print("'tb_traffic' 생성")
        
        cursor.execute("SHOW TABLES LIKE 'tb_traffic';")
        result = cursor.fetchone()
        
        if result:
            print("테이블 생성 성공적")
        else:
            print("테이블 생성 실패")
except Exception as e:
    print(f"오류 발생 {e}")
    
finally:
    if connection and connection.open:
        connection.close()
        print("연결 성공적 종료")
    else:
        print("이미 종료 됨")
```

데이터 삽입

```python
import os
import json
import pymysql

connection = pymysql.connect(
    host=,
    port=, 
    user=,    
    password=,
    database=,  
    charset=
)

directory_path = r'C:\Users\ran14\Workspace\park-to-play Workspace\db 코드\traffic'

# json 리스트 가져오기

try:
    json_files = [f for f in os.listdir(directory_path) if f.endswith('.json')]
    print(f"json 파일 {len(json_files)}개")
except FileNotFoundError:
    print(f"경로를 찾을 수 없음")
    json_files = []
    

# 데이터 삽입 SQL 쿼리(traffic_id는 자동 부여하므로 제외)
insert_query = """
INSERT INTO tb_traffic (name, congestion, distance, firstUpdateTime, speed, direction, coordinates)
VALUES (%s, %s, %s, %s, %s, %s, %s)
"""

try:
    with connection.cursor() as cursor:
        for json_file in json_files:
            file_path = os.path.join(directory_path, json_file) #  각 파일의 전체 경로
            print(f"파일 불러오는 중 : {file_path}")
            
            # json 읽기
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
                
                # 데이터가 리스트인지 확인
                if isinstance(data, list):
                    for record in data:
                        cursor.execute(
                            insert_query,
                            (
                                record['name'],
                                record['congestion'],
                                str(record['distance']),
                                record['firstUpdateTime'],
                                record['speed'],
                                record['direction'],
                                record['coordStr']
                            )
                        )
                # 단일 데이터일 경우 처리
                else:
                    cursor.execute(
                        insert_query,
                        (
                            data['name'],
                            data['congestion'],
                            str(data['distance']),
                            data['firstUpdateTime'],
                            data['speed'],
                            data['direction'],
                            data['coordStr']
                        )
                    )
        # 모두 삽입 후 커밋
        connection.commit()
        print("모든 데이터 성공적으로 삽입")
                    
except Exception as e:
    print(f"오류 발생{e}")
    
finally:
    connection.close()
```
