```python
from sqlalchemy import create_engine
import pandas as pd

# SQLAlchemy 엔진 생성 (MySQL 연결 문자열 사용)
engine = create_engine("mysql+pymysql://", echo=False)

# 데이터베이스에서 SQL 쿼리를 사용해 데이터 불러오기
try:
    # SQL 쿼리 실행
    parking_query = "SELECT * FROM tb_parking"  # 주차장
    weather_query = "SELECT * FROM tb_weather"  # 날씨
    traffic_query = "SELECT * FROM tb_traffic"  # 교통현황
    # gis_query = "SELECT * FROM tb_gis" # gis

    # pandas의 read_sql을 사용하여 SQL 쿼리 결과를 DataFrame으로 변환
    parking_df = pd.read_sql(parking_query, engine)
    weather_df = pd.read_sql(weather_query, engine)
    traffic_df = pd.read_sql(traffic_query, engine)
    # gis_df = pd.read_sql(gis_query, engine)

finally:
    # SQLAlchemy 엔진은 자동으로 연결을 관리하므로 명시적으로 종료할 필요는 없습니다.
    pass

# 불러온 데이터 확인
print(parking_df.head())
print(weather_df.head())
print(traffic_df.head())
# print(gis_df.head())
```

## 주차장 데이터 전처리

```python
import numpy as np
import holidays

def preprocess_parking(parking_df):
    parking_df.drop(columns={'lat', 'lng', 'park_phone', 'park_address', 'count', 'date', 'day_maximum', 'fulltime_monthly',
                         'weekday_begin_time', 'weekday_end_time', 'weekend_begin_time', 'weekend_end_time', 'holiday_begin_time', 'holiday_end_time'}, inplace=True)
    # parking_list에서 주차장 이름만 추출
    parking_names = [item[0] for item in parking_list]
    # parking_df에서 park_name이 parking_names에 포함된 경우 필터링
    parking_df = parking_df[parking_df['park_name'].isin(parking_names)]

    # date 열을 기준으로 시간별로 정렬
    parking_df = parking_df.sort_values(by='cur_parking_time')

    parking_df['cur_parking_new'] = np.nan
    parking_df['occupancy_rate'] = np.nan
    parking_df['purpose'] = np.nan


    # 유효한 데이터프레임에서 'cur_parking_time'을 DateTime 형식으로 변환
    parking_df['cur_parking_time'] = pd.to_datetime(parking_df['cur_parking_time'])
    parking_df['weekday'] = parking_df['cur_parking_time'].dt.weekday
    # 주말이면 1 그냥 평일이면 0의 값을 가지는 'holiday' 칼럼 생성
    parking_df['holiday'] = parking_df['weekday'].apply(lambda x: 1 if x >= 5 else 0)
    # 평일 중 공휴일도 1로 변경
    kr_holidays = holidays.KR(years=2024)
    holiday_dates = set(kr_holidays.keys())
    parking_df['holiday'] = parking_df.apply(
        lambda row: 1 if row['cur_parking_time'] is not pd.NaT and row['cur_parking_time'].date() in holiday_dates else row['holiday'],
        axis=1)



    for item in parking_list:
        park = item[0]
        purpose_value = item[1]

        # 주차장 이름이 일치하는 경우 purpose 값 업데이트
        parking_df.loc[parking_df['park_name'] == park, 'purpose'] = purpose_value

        # 각 주차장별로 추가 전처리 수행
        episode_data = parking_df[parking_df['park_name'] == park].copy()

        # 여의도 사학연금(TP타워)일 때 cur_parking 값에서 494 차감
        if park == '여의도 사학연금(TP타워)':
            episode_data['cur_parking'] = episode_data['cur_parking'] - 494
            episode_data['available_parking'] = episode_data['available_parking'] + 494

        # 코레일유통 본사 사옥 주차장일 때 capacity 및 available_parking 값을 증가
        if park == '코레일유통 본사 사옥 주차장':
            episode_data['capacity'] = episode_data['capacity'] + 40
            episode_data['available_parking'] = episode_data['available_parking'] + 40

        # 당산근린공원 공영(구)일 때 capacity 차감 및 available_parking 값을 증가
        if park == '당산근린공원 공영(구)':
            episode_data['cur_parking'] = episode_data['cur_parking'] - 43
            episode_data['available_parking'] = episode_data['available_parking'] + 43

        episode_data = episode_data.reset_index(drop=True)

        # 점유율 계산 (cur_parking/capacity * 100)
        episode_data['occupancy_rate'] = ((episode_data['cur_parking']) / (episode_data['capacity'])) * 100
        episode_data['occupancy_rate_cleaned'] = episode_data['occupancy_rate'].replace(100, np.nan).interpolate()
        episode_data['occupancy_rate_cleaned'] = episode_data['occupancy_rate_cleaned'].replace(0, np.nan).interpolate()
        episode_data['change'] = episode_data['occupancy_rate_cleaned'].diff()

        max_change_row = episode_data.loc[episode_data['change'].abs().idxmax()]
        print(max_change_row)
        max_change_value = abs(max_change_row['change'])

        episode_data.drop(columns={'change'}, inplace=True)
        episode_data['change'] = episode_data['occupancy_rate'].diff()

        for i in range(1, len(episode_data)):
            if (episode_data.loc[i, 'occupancy_rate'] == 100) | (episode_data.loc[i, 'occupancy_rate'] == 0):
                if episode_data.loc[i, 'que_status'] == "2시간":
                    episode_data.loc[i, 'occupancy_rate'] = np.nan
                if (abs(episode_data.loc[i, 'change']) > max_change_value) | (pd.isna(episode_data.loc[(i-1), 'occupancy_rate'])):
                    episode_data.loc[i, 'occupancy_rate'] = np.nan

        episode_data['occupancy_rate'] = episode_data['occupancy_rate'].interpolate()
        episode_data['cur_parking_new'] = episode_data['occupancy_rate'] * episode_data['capacity'] / 100

        # 원래 parking_df에 전처리된 데이터 병합 (cur_parking, available_parking, capacity, cur_parking_new, occupancy_rate)
        temp_df = episode_data[['park_name', 'cur_parking_time', 'cur_parking', 'available_parking', 'capacity', 'cur_parking_new', 'occupancy_rate', 'rates', 'time_rate', 'add_rates']]

        for index, row in temp_df.iterrows():
            mask = (parking_df['park_name'] == row['park_name']) & (parking_df['cur_parking_time'] == row['cur_parking_time'])
            parking_df.loc[mask, 'cur_parking'] = row['cur_parking']
            parking_df.loc[mask, 'available_parking'] = row['available_parking']
            parking_df.loc[mask, 'capacity'] = row['capacity']
            parking_df.loc[mask, 'cur_parking_new'] = row['cur_parking_new']
            parking_df.loc[mask, 'occupancy_rate'] = row['occupancy_rate']

    return parking_df
```

```python
import pandas as pd

def preprocess_weather(weather_df):
    # 사용할 컬럼 선택
    selected_features = [
        'address', 'awsStnName', 'awsPty', 'awsPcpM15', 'awsPcpHr1', 'awsPcpHr12', 'awsPcpDay',
        'awsTmp', 'awsWss', 'awsWs1', 'awsWs10', 'awsReh', 'tm'
    ]

    weather_df = weather_df[selected_features]
    # 컬럼명 변경
    rename_columns = {
        'address': '주소',
        'awsStnName': '관측소 이름',
        'awsPty': '강수 형태',
        'awsPcpM15': '15분 강수량',
        'awsPcpHr1': '1시간 강수량',
        'awsPcpHr12': '12시간 강수량',
        'awsPcpDay': '하루 강수량',
        'awsTmp': '기온',
        'awsWss': '순간 최대 풍속',
        'awsWs1': '1분 평균 풍속',
        'awsWs10': '10분 평균 풍속',
        # 'awsChillTmp': '체감 온도',
        # 'awsWds': '순간 최대 풍향',
        # 'awsWd1': '1분 평균 풍향',
        # 'awsWd10': '10분 평균 풍향',
        # 'awsTd': '이슬점 온도',
        'awsReh': '습도',
        'tm': '관측 시간',
    }
    weather_df = weather_df.rename(columns=rename_columns)

    # 주소 필터링 (영등포구만)
    weather_df = weather_df[weather_df['주소'].str.startswith('서울특별시 영등포구')]

    # 강수 형태 인코딩
    weather_df['강수 형태'] = weather_df['강수 형태'].map({'무': 0, '유': 1}).astype(bool)

    # '관측 시간' 컬럼을 datetime 형식으로 변환
    weather_df['관측 시간'] = pd.to_datetime(weather_df['관측 시간'], format='%Y%m%d%H%M', errors='coerce')

    # 관측소 이름을 단순화
    weather_df['관측소 이름'] = weather_df['관측소 이름'].replace(r'^영등포.*', '영등포', regex=True)

    # 불필요한 컬럼 제거 (주소 컬럼 삭제)
    weather_df.drop(columns=['주소'], inplace=True)

    # 강수량 여부를 True/False로 변환
    weather_df.loc[weather_df['15분 강수량'].notna(), '15분 강수량'] = True
    weather_df.loc[weather_df['15분 강수량'].isna(), '15분 강수량'] = False
    weather_df.loc[weather_df['1시간 강수량'].notna(), '1시간 강수량'] = True
    weather_df.loc[weather_df['1시간 강수량'].isna(), '1시간 강수량'] = False
    weather_df.loc[weather_df['12시간 강수량'].notna(), '12시간 강수량'] = True
    weather_df.loc[weather_df['12시간 강수량'].isna(), '12시간 강수량'] = False
    weather_df.loc[weather_df['하루 강수량'].notna(), '하루 강수량'] = True
    weather_df.loc[weather_df['하루 강수량'].isna(), '하루 강수량'] = False
    weather_df = weather_df.sort_values(by='관측 시간')
    weather_df = weather_df.drop_duplicates(subset=['관측소 이름', '관측 시간'])

    return weather_df
```

## 주차장, 날씨, gis데이터 병합

```python
def merge(parking_df, weather_df, gis_df):
    # 'cur_parking_time'을 DateTime 형식으로 변환 및 분 단위로 반올림
    parking_df['cur_parking_time'] = pd.to_datetime(parking_df['cur_parking_time'])
    parking_df['cur_parking_time'] = parking_df['cur_parking_time'].dt.floor('T')

    # 'target_time' 생성 (현재 시간에서 20분 전 시간)
    parking_df['target_time'] = parking_df['cur_parking_time'] - pd.Timedelta(minutes=20)

    # weather_df에서 '관측 시간'을 DateTime 형식으로 변환
    weather_df['관측 시간'] = pd.to_datetime(weather_df['관측 시간'])

    # A 리스트의 주차장은 한강, B 리스트의 주차장은 영등포로 매핑
    parking_df['관측소 이름'] = parking_df['park_name'].apply(lambda x: '한강' if x in hanriver else ('영등포' if x in yeongdongpo else None))

    # parking_df와 gis_df를 'park_name'과 '주차장 이름'을 기준으로 병합
    parking_df = pd.merge(parking_df, gis_df, left_on='park_name', right_on='주차장 이름', how='left')

    # 'target_time'과 '관측소 이름'을 기준으로 parking_df와 weather_df 병합
    df_merged = pd.merge(parking_df, weather_df, left_on=['관측소 이름', 'target_time'], right_on=['관측소 이름', '관측 시간'], how='left')

    # 필요한 열만 선택해서 반환
    return df_merged
```

## 혼잡도 계산

```python
def cal_confusion(df):
    df.drop(columns={'traffic_id', 'roadName', 'roadDrcType', 'linkNo', 'startNodeId', 'endNodeId', 'travelTime'}, inplace=True)
    link_id = [item[1] for item in linkid_list]
    df = df[df['linkId'].isin(link_id)]
    for park_name, link_id, max_spd in linkid_list:
        # df에서 'linkId'가 현재 link_id와 일치하는 행 필터링
        df.loc[df['linkId'] == link_id, 'park_name'] = park_name
        df.loc[df['linkId'] == link_id, 'max_spd'] = max_spd
    df = df[['park_name', 'linkId', 'speed', 'max_spd', 'createdDate']]
    df['confusion'] = df['speed'] / df['max_spd']

    # confusion 열의 평균을 주차장 이름 별로 계산
    confusion_df = df.groupby(['park_name', 'createdDate'])['confusion'].mean().reset_index()
    confusion_df.loc[confusion_df['confusion'] >= 1, 'confusion'] = 1.00
    confusion_df['confusion'] = 1 - confusion_df['confusion']
    confusion_df['confusion'] = confusion_df['confusion'].round(2)
    confusion_df['createdDate'] = pd.to_datetime(confusion_df['createdDate'], format='%Y%m%d%H%M%S')
    confusion_df = confusion_df.sort_values(by='createdDate')

    return confusion_df
```

```python
def get_confusion(row, traffic_df):
    # cur_parking_time을 datetime으로 변환 (만약 이미 datetime 형식이라면 이 줄은 생략 가능)
    row['cur_parking_time'] = pd.to_datetime(row['cur_parking_time'])

    # traffic_df에서 createdDate가 datetime 형식인지 확인하고, 변환 필요 시 변환
    traffic_df['createdDate'] = pd.to_datetime(traffic_df['createdDate'])

    # 동일한 park_name을 가진 traffic_df 데이터 필터링
    matched_traffic = traffic_df[
        (traffic_df['park_name'] == row['park_name']) &
        (row['cur_parking_time'] - pd.Timedelta(minutes=25) <= traffic_df['createdDate']) &
        (traffic_df['createdDate'] <= row['cur_parking_time'] - pd.Timedelta(minutes=5))
    ]

    # confusion 값을 반환, 없으면 NaN 반환
    if not matched_traffic.empty:
        return matched_traffic['confusion'].values[0]
    else:
        return None
```
