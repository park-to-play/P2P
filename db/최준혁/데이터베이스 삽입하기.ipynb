```python
import pymysql
import pandas as pd

# 데이터베이스 접속 정보
connection = pymysql.connect(
    host='',
    port=,
    database="",
    user="",
    password="",
    charset="utf8"
)

# 샘플로 사용할 CSV 파일 경로
csv_file = '/Users/sopung/Desktop/workspace/address/address_folder/output_with_coordinates_1.csv'

# SQL INSERT 쿼리
insert_query = """
    INSERT INTO tb_address (address, zipcode, building_name, lat, lng) 
    VALUES (%s, %s, %s, %s, %s)
"""

try:
    with connection.cursor() as cursor:
        # CSV 파일에서 첫 번째 행을 읽어옴
        df = pd.read_csv(csv_file, encoding='utf-8')

        # 첫 번째 행의 데이터를 추출하고, NaN 값을 None으로 변환
        row = df.iloc[0]
        address = row['address']
        zipcode = row['zipcode'] if pd.notna(row['zipcode']) else None
        building_name = row['building_name'] if pd.notna(row['building_name']) else None
        lat = row['latitude'] if pd.notna(row['latitude']) else None
        lng = row['longitude'] if pd.notna(row['longitude']) else None

        # 데이터 삽입
        cursor.execute(insert_query, (address, zipcode, building_name, lat, lng))

        # 변경 사항 커밋
        connection.commit()
        print(f"샘플 데이터 삽입 성공: {address}, {zipcode}, {building_name}, {lat}, {lng}")

except Exception as e:
    print(f"오류 발생: {e}")
    connection.rollback()

finally:
    # 연결 종료
    connection.close()
```

```python
import pymysql
import pandas as pd
import glob

# 데이터베이스 접속 정보
connection = pymysql.connect(
    host='',
    port=,
    database="",
    user="",
    password="",
    charset="utf8"
)

# 파일 불러오기 (output_with_coordinates_1.csv ~ output_with_coordinates_54.csv)
csv_files = glob.glob('/Users/sopung/Desktop/workspace/address/address_folder/output_with_coordinates_*.csv')

# SQL INSERT 쿼리
insert_query = """
    INSERT INTO tb_address (address, zipcode, building_name, lat, lng) 
    VALUES (%s, %s, %s, %s, %s)
"""

# 배치 크기 정의
BATCH_SIZE = 1000

try:
    with connection.cursor() as cursor:
        for file in csv_files:
            # CSV 파일을 DataFrame으로 불러오기
            df = pd.read_csv(file, encoding='utf-8')

            # NaN 값을 None으로 변환 (여기서 value=None을 사용합니다)
            df = df.where(pd.notnull(df), None)

            # 데이터프레임을 튜플 리스트로 변환
            data_to_insert = [
                (row['address'], row['zipcode'], row['building_name'], row['latitude'], row['longitude'])
                for _, row in df.iterrows()
            ]
            
            # 데이터를 배치 크기(BATCH_SIZE)로 나누어서 INSERT
            for i in range(0, len(data_to_insert), BATCH_SIZE):
                batch = data_to_insert[i:i + BATCH_SIZE]
                cursor.executemany(insert_query, batch)
                print(f"{len(batch)}개의 행을 삽입했습니다.")

        # 데이터베이스에 변경 사항 적용 (커밋)
        connection.commit()
        print("모든 데이터를 성공적으로 삽입했습니다.")

except Exception as e:
    print(f"오류 발생: {e}")
    connection.rollback()

finally:
    # 연결 종료
    connection.close()
```
